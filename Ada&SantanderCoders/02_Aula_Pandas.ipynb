# Séries

```python
# Cria uma lista
lista = [12, 23, 34, 56, 67]
```

## pd.Series()

```python
# Cria uma série
series = pd.Series(lista)
print(series)
```

```python
# Seleciona um item da série, conforme index informado como argumento
print(series[0])
```

```python
# Cria um dicionário com nome e notas
notas = {'Tiago':8, 'Guilherme':9.5, 'João':10, 'Ana':7.5}

# Cria uma série
series_notas = pd.Series(notas)
print(series_notas)
```

```python
# Seleciona um item da série, conforme chame informada como argument
print(series_notas['Guilherme'])
```

## sum()

```python
# Soma os valores da série
series_notas.sum()
```

## mean()

```python
# Calcula média
series_notas.mean()
```

## median()

```python
# Calcula a média
series_notas.median()
```

## describe()

```python
# Mostra a estatísticas descritivas da série
series_notas.describe()
```

# DataFrame

```python
# Cria uma dicionário com autores, preços e títulos
dicionario = {
    'Autores':['Rick Riordan', 'J. R. R. Tolkien', 'Rick Riordan', 'Machado de Assis'],
    'Títulos':['O Ladrão de Raios', 'A Sociedade do Anel', 'Mar de Monstros', 'Memórias Postumas de Brás Cubas'],
    'Preços':[41.2, 35.7, 39.5, 40.5]}
```

## pd.DataFrame()

```python
# Cria uma dataframe
df = pd.DataFrame(dicionario)
```

```python
# Exibe a coluna
df['Autores']
```

```python
# Exibe apenas o valor especificado do index
df['Autores'][2]
```

```python
# Soma a coluna selecionada
df['Preços'].sum()
```

```python
# Cria uma lista validando os elementos que pode ser usado para filtragem
mascara = (df['Autores'] == 'Rick Riordan')
mascara
```

```python
# Exibe os elementos que atendem os critérios de filtragem especificados acima
df[mascara]
```

## pd.concat()

```python
# Cria uma nova linha e concatena com o dataframe
nova_linha = pd.DataFrame({'Autores': ['Rick Riordan'], 'Títulos': ['A Maldicao do Titãn'], 'Preços': [40.2]})
df = pd.concat([df, nova_linha], ignore_index=True)
df
```

```python
# Seleciona o elemento específico para correção
selecao = df['Títulos'] == 'A Maldicao do Titãn'
selecao
```

```python
# Seleciona o elemento que atende o critério
df.loc[selecao]
```

```python
# Seleciona dentro da seleção a variável títulos
df.loc[selecao, 'Títulos']
```

```python
# Atribui um novo valor ao campo selecionado
df.loc[selecao, 'Títulos'] = 'A Maldição do Titãn'
df
```

# Base de dados: Taxi

```python
import seaborn as sns

# Carregar o conjunto de dados diretamente da biblioteca Seaborn, sem precisar baixar ou armazenar o arquivo localmente
taxis = sns.load_dataset("taxis")
```

```python
taxis.head()
```

## count()

```python
# Conta o número de registros para todas as colunas
taxis.count()
```

```python
# Opção 1
# Conta o número de registros para culuna selecionada
taxis.pickup_borough.count()
```

```python
# Opção 1
# Conta o número de registros para culuna selecionada
taxis['pickup_borough'].count()
```

## value_counts()

Conta a frequência que cada valor aparece.

```python
# Opção 1
taxis.pickup_borough.value_counts()
```

```python
# Opção 2
taxis['pickup_borough'].value_counts()
```

```python
# Opção 3
taxis.value_counts('pickup_borough')
```

## groupby() e agg()

* groupby: separa os dados em grupos com base em uma ou mais variáveis.
* agg: aplica várias funções de agregação aos dados de cada grupo.

```python
# Opção 1
taxis.groupby(['pickup_borough']).tip.agg(['min', 'mean', 'median', 'max'])
```

```python
# Opção 2
taxis.groupby(['pickup_borough'])['tip'].agg(['min', 'mean', 'median', 'max'])
```

```python
# Opção 3
taxis.tip.groupby(taxis.pickup_borough).agg(['min', 'mean', 'median', 'max'])
```

```python
# Opção 4
taxis.tip.groupby(taxis['dropoff_borough']).agg(['min', 'mean', 'median', 'max'])
```
