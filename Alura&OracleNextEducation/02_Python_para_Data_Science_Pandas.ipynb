# 3.1 Criando dicionários

Listas são coleções sequenciais, isto é, os itens destas sequências estão ordenados e utilizam índices (números inteiros) para acessar os valores.

Os dicionários são coleções um pouco diferentes. São estruturas de dados que representam um tipo de mapeamento. Mapeamentos são coleções de associações entre pares de valores onde o primeiro elemento do par é conhecido como chave (*key*) e o segundo como valor (*value*).

```
dicionario = {key_1: value_1, key_2: value_2, ..., key_n: value_n}
```

https://docs.python.org/3.6/library/stdtypes.html#typesmapping

````python
# Cria uma lista com marcas de carros.
carros = ['Jetta Variant', 'Passat', 'Crossfox']
carros
````

````python
# Cria uma lista com valores correspondentes aos carros da lista acima.
valores = [88078.64, 106161.94, 72832.16]
valores
````

````python
# Informa a posição do elemento na lista.
carros.index('Passat')
````

````python
# Informa o valor corresponte ao item de mesma posição na lista valor.
valores[carros.index('Passat')]
````

````python
# Cria um dicionário que recebe, chave e valor.
dados = {'Jetta Variant': 88078.64, 'Passat': 106161.94, 'Crossfox': 72832.16}
dados
````

````python
# Informa o tipo de dado.
type(dados)
````

## Exercício - Quilometragem média de um veículo

Cálcule da quilometragem média que um veículo rodou por ano. Não existe uma função pronta para o cálculo desta estatística em Python, e por isso precisamos escrever uma função que obtenha este valor.

Considere o conjunto de informações e a fórmula abaixo para responder o problema:

image.png

````python
# Dicionário com os dados para resolução do problema acima.
dados_km = {
    'Crossfox': {'km': 35000, 'ano': 2005},
    'DS5': {'km': 17000, 'ano': 2015},
    'Fusca': {'km': 130000, 'ano': 1979},
    'Jetta': {'km': 56000, 'ano': 2011},
    'Passat': {'km': 62000, 'ano': 1999}
}
````

````python
# Resolução.
# Cria uma função para calcular quilometragem média.
def km_media(dataset, ano_atual):
    for item in dataset.items():
        result = item[1]['km'] / (ano_atual - item[1]['ano'])
        print(result)
````

### Funções que retornam mais de um valor

#### Formato padrão

```
def <nome>(<param_1>, <param_2>, ..., <param_n>):
    <instruções>
    return (<resultado_1>, <resultado_2>, ..., <resultado_n>)
```

````python
# Cria uma função que recebe como parâmetro um listas.
def media(lista):
  valor = sum(lista) / len(lista)
  return (valor, len(lista)) # Retorna dois valores.
````

````python
# Executa a função, ajustada acima.
media([1, 2, 3, 4, 5, 6, 7, 8, 9])
````

````python
# Atribui os valores as variáveis conforme sequência, ou seja, faz o desempacotamento da tupla criada pela função.
resultado, n = media([1, 2, 3, 4, 5, 6, 7, 8, 9])
````

A estrutura a seguir é a definição de uma função que calcula as quilometragens médias anuais de cada veículo e retorna um dicionário com os nomes dos veículos como chaves e as quilometragens médias como valores:

## Exercício - Elaborando um pouco mais a nossa função

Uma elaboração extra, que também pode ser interessante, é a atualização do próprio input da função. Podemos fazer com que a nossa função retorne as informações do dicionário dados, incluindo as informações de quilometragem média anual.

A estrutura a seguir é a definição de uma função que calcula as quilometragens médias anuais de cada veículo, atualiza o dicionário de entrada e retorna este dicionário:

````python
# Dicionário com os dados para resolução do problema acima.
dados_km = {
    'Crossfox': {'km': 35000, 'ano': 2005},
    'DS5': {'km': 17000, 'ano': 2015},
    'Fusca': {'km': 130000, 'ano': 1979},
    'Jetta': {'km': 56000, 'ano': 2011},
    'Passat': {'km': 62000, 'ano': 1999}
}
````

````python
# Resolução.
# Cria uma função para calcular quilometragem média, e atualiza o dicionário com o valor calculado.
def km_media(dataset, ano_atual):
    result = {}
    for item in dataset.items():
        media = item[1]['km'] / (ano_atual - item[1]['ano'])
        item[1].update({ 'km_media': media })
        result.update({ item[0]: item[1] })
    return result
````

```
    s = pd.Series(dados, index = index)

O argumento *dados* pode ser um dicionário, uma lista, um array Numpy ou uma constante.

### DataFrames

DataFrame é uma estrutura de dados tabular bidimensional com rótulos nas linha e colunas. Como a Series, os DataFrames são capazes de armazenar qualquer tipo de dados.


```
    df = pd.DataFrame(dados, index = index, columns = columns)
```

O argumento *dados* pode ser um dicionário, uma lista, um array Numpy, uma Series e outro DataFrame.
