Uma pessoa criou um notebook em inglês para analisar a quantidade de pessoas em diferentes capitais do mundo.

Porém, antes de começar suas respectivas análises, decidiu traduzir as colunas e os campos para o português, conforme ilustram as imagens abaixo:

image.png

Renomei as colunas e campos.

```
import pandas as pd
```

```
population = (
{
'Country': ['Belgium', 'India', 'Japan'],
'Population': [12345, 67891011, 98765]
})
```

```
dados = pd.DataFrame(population)
```

```
dados
```

```
paises = {'Belgium':'Bélgica', 'India':'Índia', 'Japan':'Japão'}
dados.Country = dados.Country.map(paises)
dados
```

```
colunas = {'Country':'País', 'Population':'População'}
dados = dados.rename(columns=colunas)
dados
```
