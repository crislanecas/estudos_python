```python
# Cria uma imagem.
fig = plt.figure(figsize=(8,4)) # Personaliza o tamanho da figura.
eixo = fig.add_axes([0,0,1,1]) # Define eixo de 0 a 1 para X e Y, para usar 100% do espaço.

cores = ['red', 'blue', 'orange', 'green'] # Cria uma lista de cores.

# Cria uma visualização de caixas para todas as variáveis da coluna defida.
caixas = eixo.boxplot(df.drop('espécie', axis=1).values, # A instrução .drop() remove da plotagem a espécie, o axis=1 identifica como coluna e .values() transforma dataframe em vetor contendo somente os valores.
             patch_artist=True) # Configura cores as caixas.
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10) # Configura o título para a imagem.
eixo.set_xticklabels(df.drop('espécie', axis=1).columns) # Configura um rótulo.

# Cria um loop que irá associar cada caixa um cor da lista acima.
for caixa, cor in zip(caixas['boxes'], cores): # A função zip() permite iterar duas variáveis distintas.
    caixa.set(color=cor) # A função set() ajusta o parâmetro cor para a cor recebida no laço.

# Cria um loop que irá associar alterar os marcadores outliers ou fliers, pontos fora da caixa plotada.
for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8) # A função set() recebe o parâmetro marker='x', que altera sua visualização para um x e o markersize aumentarmos o seu tamanho.
```

# Aula 5: Customizando histogramas

```python
# Cria uma imagem.
fig = plt.figure(figsize=(5,4)) # Personaliza o tamanho da figura.
eixo = fig.add_axes([0,0,1,1]) # Define eixo de 0 a 1 para X e Y, para usar 100% do espaço.

# Calcula a média e desvio padrão, passando os valores para uma tupla.
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

# Cria uma visualização colunas com a frequência.
eixo.hist(df['comprimento_pétala'], bins=20) # Define a fonte dos dados e defini o número de colunas da imagem.
eixo.set_title('Histograma', fontsize=15, pad=10) # Configura o título para a imagem.
eixo.set_xlabel('Comprimento da pétala', fontsize=15) # Configura o label do eixo X.
eixo.grid(True) # Configura uma grade.

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma), # Faz a anotação na imagem, com formação código entre cifrões ($$), média (mu) recebe a formatação {0:.2f} onde o 1º valor define a posição da tupla e o 2º exibe duas casas decimais e \n para pularmos uma linha.
               xy=(4.5, 20), fontsize=20) # O parâmetro xy recebendo uma tupla 4.5,20 com a posição da anotação.
```

```python
# Cria uma imagem.
fig = plt.figure(figsize=(5,4)) # Personaliza o tamanho da figura.
eixo = fig.add_axes([0,0,1,1]) # Define eixo de 0 a 1 para X e Y, para usar 100% do espaço.

# Calcula a média e o desvio padrão, passando os valores para uma tupla.
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

# Cria uma visualização colunas com a frequência.
eixo.hist(df['comprimento_pétala'], bins=20) # Define a fonte dos dados e defini o número de colunas da imagem.
eixo.set_title('Histograma', fontsize=15, pad=10) # Configura o título para a imagem.
eixo.set_xlabel('Comprimento da pétala', fontsize=15) # Configura o label do eixo X.
eixo.grid(True) # Configura uma grade.

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), # Faz a anotação na imagem, com formação do código entre cifrões ($$), onde média (mu) recebe a formatação {0:.2f}, o 1º valor define a posição da tupla e o 2º exibe duas casas decimais e \n para pularmos uma linha.
               xy=(4.5, 20), fontsize=20) # O parâmetro xy recebendo uma tupla 4.5,20 com a posição da anotação.

# Adiciona uma linha vertical indicando o valor médio, calculado acima.
eixo.axvline(mu, color = 'k', linestyle='--') # Define cor e estilo da reta.
eixo.annotate('média', xy=(mu-1.3, 28), fontsize=20) # Adiciona uma anotação na posição xy.

# Adiciona uma linha vertical indicando o valor da mediana.
eixo.axvline(df['comprimento_pétala'].median(), color='r', linestyle='--') # Calcula a mediana e define cor e estilo da reta.
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median()+0.1, 31), fontsize=20, color='r') # Adiciona uma anotação na posição xy.
```

### Histograma Iris Versicolor

```python
# Cria uma imagem.
fig = plt.figure(figsize=(5,4)) # Personaliza o tamanho da figura.
eixo = fig.add_axes([0,0,1,1]) # Define eixo de 0 a 1 para X e Y.

# Filtra os dados, selecionando apenas os registros cujas espécies são Iris-versicolor.
df_iv = df[df['espécie'] == 'Iris-versicolor']

# Calcula a média e o desvio padrão, passando os valores para uma tupla.
mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

# Cria uma visualização colunas com a frequência.
eixo.hist(df_iv['comprimento_pétala'], bins=20) # Define a fonte dos dados e defini o número de colunas da imagem.
eixo.set_title('Iris Versicolor', fontsize=15, pad=10) # Configura o título para a imagem.
eixo.set_xlabel('Comprimento da pétala', fontsize=15)  # Configura o label do eixo X.
eixo.grid(True) # Configura uma grade.

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), # Faz a anotação na imagem, com formação para os valores da tupla.
               xy=(4.6, 6), fontsize=20) # O parâmetro xy recebendo uma tupla com a posição da anotação.

# Adiciona uma linha vertical indicando o valor médio, calculado acima.
eixo.axvline(mu, color = 'k', linestyle='--') # Define cor e estilo da reta.
eixo.annotate('média', xy=(mu-0.45, 5.5), fontsize=20)

# Adiciona uma linha vertical indicando o valor da mediana.
eixo.axvline(df_iv['comprimento_pétala'].median(), color='r', linestyle='--') # Calcula a mediana e define cor e estilo da reta.
eixo.annotate('mediana', xy=(df_iv['comprimento_pétala'].median()-0.6, 6.5), fontsize=20, color='r') # Adiciona uma anotação na posição xy.

# Salva a imagem, com o nome escolhido e extensão png.
fig.savefig('histograma_iv.png', bbox_inches='tight') # O argumento tight, reduz a área em volta.
```

### Histograma combinado

```python
# Cria uma imagem combinada.
largura, altura = Image.open('histograma_ivc.png').size # Define largura e altura da imagem
combinada = Image.new("RGB", (3*largura, altura)) # Combina uma nova imagem com 3 vezes a largura e altura da imagem passada acima como parâmetro.
intervalo = 0 # Define o intervalo entre as imagens.
for imagem in map(Image.open, ['histograma_iv.png', # Mapeia as imagens listadas como parâmetro.
                               'histograma_is.png',
                               'histograma_ivc.png']):
    combinada.paste(imagem, (intervalo, 0)) # Define a posição para colar as imagens.
    intervalo += largura # Encrementa o intervalo com += largura, o valor da próprio somado à largura da nova imagem.

# Salva a nova imagem.
combinada.save('combinada.png')
```
