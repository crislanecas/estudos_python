```python
# Importa a biblioteca.
import pandas as pd
```

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html

```python
# Cria um dataframe.
tabela_1 = pd.DataFrame({
'Nome':['João', 'João', 'Pedro' , 'Caio'],
'Telefone': ['12121', '343434', '565656', '787878'],
'Carros': ['azul', 'preto', 'verde' , 'amarelo']})
tabela_1
```

```python
# Cria um outro dataframe.
tabela_2 = pd.DataFrame({
'Nome':['João', 'Marcelo', 'Thiago' , 'Caio'],
'Irmãos': ['1', '3', '2' , '2']})
tabela_2
```

# Inner

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'inner'.
m = pd.merge(tabela_1, tabela_2, how = 'inner', on = 'Nome')
m
```

# Outer

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'outer'.
m = pd.merge(tabela_1, tabela_2, how = 'outer')
m
```

# Right

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'right'.
m = pd.merge(tabela_2, tabela_1, how = 'right', on = 'Nome')
m
```

# Left

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'left'.
m = pd.merge(tabela_1, tabela_2, how = 'left', on = 'Nome')
m # Obtém o mesmo resultado do código acima.
```

```python
# Cria um novo dataframe.
tabela_3 = pd.DataFrame({
'Nome':['João', 'João', 'Pedro' , 'Caio'],
'Telefone': ['12121', '343434', '565656', '787878'],
'Carros': ['BRANCO', 'PRATA', 'PRATA' , 'VERMELHO']})
tabela_3
```

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'left'.
# _x Corresponde aos valores que existiam na tabela da esquerda (tabela_1).
# _y Corresponde aos valores que existiam na tabela da direita (tabela_2).
m = pd.merge(tabela_1, tabela_3, how='left', on = 'Nome')
m
```

```python
# Utiliza a função merge() para unir as duas tabelas.
# Passando como parâmetro how = 'left' e on = ['Nome', 'Telefone'], diminuindo assim a duplicidade da coluna telefone.
m = pd.merge(tabela_1, tabela_3, how='left', on = ['Nome', 'Telefone'])
m
```
